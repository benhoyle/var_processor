```python
import numpy as np
import matplotlib.pyplot as plt

from src.sources.fft import FFTSource

from src.var_processor.sensor import Sensor
from src.var_processor.stack import Stack
from src.var_processor.switch import Switch
```

```python
vec_len = 4
time_len = 4
audio = FFTSource(res_factor=2)
sensor = Sensor(audio, 4)
sensor_len = sensor.power_len
stacks = list()
for i in range(0, 5):
    stacks.append(
        Stack(sensor_len, vec_len, time_len)
    )
```

```python
switch = Switch(sensor, stacks)
```

```python
residuals, norms = switch.iterate()
```
